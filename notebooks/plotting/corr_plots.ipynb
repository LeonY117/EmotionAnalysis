```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
corr_matrix = np.genfromtxt('./corr_matrix.csv', delimiter=',')

plt.imshow(corr_matrix)

corr_sem_vad = corr_matrix[5:8, 8:]
corr_sem_ekman = corr_matrix[0:5, 8:]
corr_sem_ekman_vad = corr_matrix[0:8, 8:]

corr_sem_vad.shape
plt.imshow(corr_sem_vad)
```

```python
corr = corr_matrix

labels_ek = ['anger-disgust (Ek)', 'fear (Ek)', 'happy (Ek)', 'sad (Ek)', 'surprise (Ek)']
labels_vad = ['V', 'A', 'D']
labels_sem = ['anger (S)', 'anticipation (S)', 'disgust (S)', 'fear (S)', 'joy (S)', 'love (S)', 'optimism (S)', 'pessimism (S)', 'sadness (S)', 'surprise (S)', 'trust (S)']
labels_all = labels_ek + labels_vad + labels_sem 


fig, ax = plt.subplots(figsize=(10, 10))
# Create a heatmap of the correlation matrix

im = ax.imshow(corr, cmap='PiYG', vmin=-1, vmax=1, alpha=0.5)

ax.set_xticks(np.arange(corr.shape[1]))
ax.set_yticks(np.arange(corr.shape[0]))
ax.set_xticklabels(labels_all)
ax.set_yticklabels(labels_all)

# Rotate the ticklabels and set them to center alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        display_text = np.round(corr[i, j], 1)
        text = ax.text(j, i, display_text,
                      ha="center", va="center", color="black", fontsize=12)
        
# Set the title of the plot
ax.set_title("Label Correlation Matrix")
plt.tight_layout()

plt.savefig('corr_all.png', dpi=150)
```
